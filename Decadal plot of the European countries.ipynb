```python
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as smb
```

```python
df = pd.read_csv('/Users/Dell/Downloads/year.csv')
df.set_index('Year', inplace = True)
df.head()
```

```python
df1 = df.loc[1970:1979]
df2 = df.loc[1980:1989]
df3 = df.loc[1990:1999]
df4 = df.loc[2000:2009]
df5 = df.loc[2010:2018]

df1.head(10)
```

```python
plt.scatter(data=[df1],color=["red","blue"])
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df1, x='Year', y='EU_pm10', hue='Year', palette='deep')
plt.xlabel('Year')
plt.ylabel('particulate matter')
#plt.legend('')
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="EU_pm2.5", y='EU_pm10',hue="Year", palette='deep')
plt.xlabel('PM 2.5')
plt.ylabel('PM 10')
plt.title("Relationship between PM 10 and PM 2.5")
plt.legend('')
```

```python
plt.figure(figsize=(12,8))
sns.regplot(data=df,x="EU_pm2.5",y="EU_pm10",color="blue")
plt.title("Relationship between PM 10 and PM 2.5")
plt.xlabel("Particulate matter 2.5")
plt.ylabel("Particulate matter 10")
plt.show()
```

```python
#all the years PM 10

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Year', y='EU_pm10', hue='Year', palette='deep')
plt.xlabel('Year')
plt.ylabel('particulate matter')
plt.title("Trend Analysis of PM10 in Europe ")
plt.legend('')
```

```python
plt.figure(figsize=(12,8))
sns.regplot(data=df,x="EU_pm2.5",y="EU_pm10", hue='Year', palette='deep)
plt.title("Relationship between PM 10 and PM 2.5")
plt.xlabel("Particulate matter 2.5")
plt.ylabel("Particulate matter 10")
plt.show()
```
