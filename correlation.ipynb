```python
import pandas as pd
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv('/home/olumide/Downloads/corr/data.csv')
df.head()
```

```python
fig = plt.figure(figsize = [12,8])
sns.regplot(x='Savanna_om', y='Savanna_t2m', data = df)
plt.xlabel('Organic Matter kgm^-3')
plt.ylabel('Temperature (C)')

X = sm.add_constant(df['Savanna_om'])
model = sm.OLS(df['Savanna_t2m'], X).fit()

intercept, slope = model.params

equation = f'y = {slope:.2f}x + {intercept:.2f}'

plt.text(0.25, 30, equation, fontsize=12, color='black')


plt.show()
```

```python
fig = plt.figure(figsize = [12,8])
sns.regplot(x='Savanna_om', y='Savanna_cp', data = df)
plt.xlabel('Organic Matter kgm^-3')
plt.ylabel('Convective precipitation (mm)')

X = sm.add_constant(df['Savanna_om'])
model = sm.OLS(df['Savanna_cp'], X).fit()

intercept, slope = model.params

equation = f'y = {slope:.2f}x + {intercept:.2f}'

plt.text(0.25, 0.004, equation, fontsize=12, color='black')


plt.show()
```
